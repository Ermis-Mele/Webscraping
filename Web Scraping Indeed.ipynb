```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests, openpyxl, csv
import time
import datetime

import smtplib
```

```python
url = 'https://nl.indeed.com/jobs?q=data%20analyst&l=Amsterdam&vjk=d02a637c937cd1cc'
req = requests.get(url)
req.raise_for_status()

soup= BeautifulSoup(req.text, 'html.parser')
print(soup.prettify())

```

```python
Job_title=soup.find_all('h2', attrs={'class':'jobTitle'})
Job_title
```

```python
Job_title=soup.find_all('h2', attrs={'class':'jobTitle'})
Job_title_list=[]
for i in Job_title:
    g=i.select('span')[-1]
    Job_title_list.append(g.text)
Job_title_list
```

```python
company_name=soup.find_all('span', attrs={'class':'companyName'})
company_name_list=[]
for i in company_name:
    company_name_list.append(i.text)
company_name_list
```

```python
company_loc=soup.find_all('div', attrs={'class':'companyLocation'})
company_loc_list=[]
for i in company_loc:
    company_loc_list.append(i.text)
company_loc_list
```

```python
df= pd.DataFrame()
df['position']= Job_title_list
df['company']= company_name_list
df['company_location']= company_loc_list
df
```

```python
df.to_csv('indeed_job.csv', index=False)
```
